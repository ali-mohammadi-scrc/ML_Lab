```python
base_path = '/content/drive/MyDrive/ML_Lab'
random_dirs_params = 'random_trials_10_50'
model_names = os.listdir(base_path)
model_names = model_names[0:2] + model_names[3:] 
print(model_names)
x_test, y_test = load_cifar10(return_part='test')

input_shape = np.shape(x_test[0])
depth = 20
loss = 'categorical_crossentropy'
metrics = ['accuracy', 'MSE']

share_model = sharesnet_v1(input_shape=input_shape, depth=depth)
share_model.compile(loss=loss, metrics=metrics)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_recall_fscore_support as score

y_test_label = y_test.argmax(axis=-1)

for model_name in model_names:
    print('**** ', model_name, ' ****')

    model_dir = os.path.join(base_path, model_name, 'latest.h5')
    share_model.load_weights(model_dir)

    training_log = np.genfromtxt(os.path.join(base_path, model_name, 'training.log'), delimiter=',')[1:, 1:]
    # fig = plt.figure()
    plt.plot(training_log[:, [1, 3]])
    plt.legend(['training loss', 'validation loss'])
    plt.savefig(os.path.join(base_path, model_name, 'training_loss.png'))
    plt.show()

    print('Final', 'Training loss:', training_log[-1, 1], 'Training Acc:', training_log[-1, 0])
    print('Final', 'Validation loss:', training_log[-1, 3], 'Validation Acc:', training_log[-1, 2])
    print('Final', 'Training Err:', 1 - training_log[-1, 0], 'Test Err:', 1 - training_log[-1, 2], 'Generalization Err:', training_log[-1, 0]-training_log[-1, 2])

    y_pred = share_model.predict(x_test)
    y_pred_label = y_pred.argmax(axis=-1)

    print('Confusion Matrix', confusion_matrix(y_test_label, y_pred_label))
    print(classification_report(y_test_label, y_pred_label))
    
    precision, recall, fscore, support = score(y_test_label, y_pred_label)
    print('precision: {}'.format(precision))
    print('recall: {}'.format(recall))
    print('fscore: {}'.format(fscore))
    print('support: {}'.format(support))
```

```python
import numpy as np
import os

from ML_Lab.keras_example_cifar10_resnet.resnet import resnet_v1
from ML_Lab.keras_example_cifar10_resnet.cifar10 import load_cifar10
from ML_Lab.loss_landscape.random_direction import loss_val_filter_normalized_dir, plot_loss_val

x_test, y_test = load_cifar10(return_part='test')

input_shape = np.shape(x_test[0])
depth = 20
loss = 'categorical_crossentropy'
metrics = ['accuracy', 'MSE']
n_trials = 10
steps_res = [-1, 1, 50]

base_path = '/content/drive/MyDrive/ML_Lab'
model_base_path = '/content/drive/MyDrive/ML_Lab'

share_model = sharesnet_v1(input_shape=input_shape, depth=depth)
share_model.compile(loss=loss, metrics=metrics)
print(share_model.summary())

model = resnet_v1(input_shape=input_shape, depth=depth)
#       model.set_weights(w)
model.compile(loss=loss, metrics=metrics)
print(model.summary())
total_models = len(os.listdir(model_base_path))

for i in range(total_models):
    if(i>7):
      model_name = os.listdir(model_base_path)[i]
      print('**** ', model_name, ' ****')

      model_dir = os.path.join(model_base_path, model_name, 'latest.h5')
      save_dir = os.path.join(
          base_path, model_name, 'random_trials_' + str(n_trials)+'_'+str(steps_res[-1]))

      print(model_dir, save_dir)

      if not os.path.exists(save_dir):
          os.makedirs(save_dir)

      share_model.load_weights(model_dir)
      w = []
      for model_layer in share_model.layers:
        weigh = model_layer.get_weights()
        w += list(weigh)
      w = np.array(w, dtype=np.object)
      print("W:",w.shape)
      model = resnet_v1(input_shape=input_shape, depth=depth)
      model.set_weights(w)
      model.compile(loss=loss, metrics=metrics)

      steps, evals = loss_val_filter_normalized_dir(
          model, x_test, y_test, steps_res=steps_res, n_trials=n_trials)
      np.save(os.path.join(save_dir, 'evals'), evals, allow_pickle=True)
      np.save(os.path.join(save_dir, 'steps'), steps, allow_pickle=True)
```

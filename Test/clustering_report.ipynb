```python
%cd ..
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_recall_fscore_support as score

from keras_example_cifar10_resnet.resnet import resnet_v1
from keras_example_cifar10_resnet.cifar10 import load_cifar10
```

```python
base_path = 'Results/Models'
model_names = os.listdir(base_path)


x_test, y_test = load_cifar10(return_part='test')

input_shape = np.shape(x_test[0])
depth = 20
loss = 'categorical_crossentropy'
metrics = ['accuracy', 'MSE']

model = resnet_v1(input_shape=input_shape, depth=depth)
model.compile(loss=loss, metrics=metrics)
```

```python
y_test_label = y_test.argmax(axis=-1)

for model_name in model_names:
    print('**** ', model_name, ' ****')

    model_dir = os.path.join(base_path, model_name, 'latest.h5')
    model.load_weights(model_dir)

    training_log = np.genfromtxt(os.path.join(base_path, model_name, 'training.log'), delimiter=',')[1:, 1:]
    plt.plot(training_log[:, [1, 3]])
    plt.show()
    print('Final', 'Training loss:', training_log[-1, 1], 'Training Acc:', training_log[-1, 0])
    print('Final', 'Validation loss:', training_log[-1, 3], 'Validation Acc:', training_log[-1, 2])
    print('Final', 'Training Err:', 1 - training_log[-1, 0], 'Test Err:', 1 - training_log[-1, 2], 'Generalization Err:', training_log[-1, 0]-training_log[-1, 2])

    y_pred = model.predict(x_test)
    y_pred_label = y_pred.argmax(axis=-1)

    print('Confusion Matrix', confusion_matrix(y_test_label, y_pred_label))
    print(classification_report(y_test_label, y_pred_label))
    
    precision, recall, fscore, support = score(y_test_label, y_pred_label)
    print('precision: {}'.format(precision))
    print('recall: {}'.format(recall))
    print('fscore: {}'.format(fscore))
    print('support: {}'.format(support))
```

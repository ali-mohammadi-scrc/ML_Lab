```python
%cd ..
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import os

from loss_landscape.random_direction import plot_loss_val
```

```python
base_path = 'Results/Resnet/Models'
model_names = os.listdir(base_path)
```

```python
random_dirs_params = 'random_trials_20_50'

for model_name in model_names:
    print('**** ', model_name, ' ****')

    steps = np.load(os.path.join(base_path, model_name, random_dirs_params, 'steps.npy'), allow_pickle=True)
    evals = np.load(os.path.join(base_path, model_name, random_dirs_params, 'evals.npy'), allow_pickle=True)
    evals = evals[:, :, [2, 1]] # Acc - MSE
    # evals = evals[:, :, [0, 1]]
    evals_mean = np.mean(evals, 0)

    fig, ax_loss = plt.subplots()
    ax_acc = ax_loss.twinx()
    plt_axes=(ax_loss, ax_acc)

    plot_loss_val(steps=steps, trials_evals=evals, plt_axes=plt_axes)
    plot_loss_val(steps=steps, trials_evals=[evals_mean], plt_axes=plt_axes, plt_colors=('g', 'y'), linestyle='-')
    ax_loss.set_ylabel('loss', size="x-large")
    ax_acc.set_ylabel('accuracy', size="x-large")
    plt.show()
```

```python
%matplotlib widget

base_path = 'Results/Resnet/Models'
random_dirs_params = 'random_trials_50_2d'

fig = plt.figure(figsize=4*plt.figaspect(1))
model_names = np.array(os.listdir(base_path))[[-1, -2, 0, 4, 5, 6, 1, 2, 3]]

for i, model_name in enumerate(model_names):
    print('**** ', model_name, ' ****')

    steps = np.load(os.path.join(base_path, model_name, random_dirs_params, 'steps.npy'), allow_pickle=True)
    evals = np.load(os.path.join(base_path, model_name, random_dirs_params, 'evals.npy'), allow_pickle=True)
    evals = evals[:, :, [2, 1]] # Acc - MSE
    # evals = evals[:, :, [0, 1]]


    X, Y = np.meshgrid(steps, steps)
    Z = evals[:, :, 0]
    Z = np.clip(Z, 0, 50)

    # fig = plt.figure()
    # ax = Axes3D(fig)
    ax = fig.add_subplot(3, 3, i+1, projection='3d')
    ax.set_title(model_name, fontsize=20)
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=True)
    # ax.plot_wireframe(X, Y, Z, rstride=5, cstride=5)
    # fig.colorbar(surf, shrink=0.5, aspect=5)

    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    # Transparent spines
    ax.w_xaxis.line.set_visible(False)
    ax.w_yaxis.line.set_visible(False)
    ax.w_zaxis.line.set_visible(False)

    # Transparent panes
    ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    

    plt.show()

fig.subplots_adjust(left=0, right=1, bottom=0, top=None, wspace=-0.4, hspace=-0.05)
# fig.subplots_adjust()
# plt.tight_layout(pad=0)
```
